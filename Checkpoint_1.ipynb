```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/MLBootcamp/netflix1.csv')
data.info()
data.head()
```

```python
data["date_added"] = pd.to_datetime(data["date_added"])
data.info()
```

```python
print(f"Number of rows: {data.shape[0]}")
print(f"Number of columns: {data.shape[1]}")
```

```python
data['year_added'] = data['date_added'].dt.year
data.head()
```

```python
types =data.groupby(['type',])[ 'type'].count().reset_index(name='count')
types=types.set_index('type')
types
```

```python
types.plot.pie(y='count',autopct='%.1f%%', shadow=True, legend= 'type' , figsize=(6,6))
plt.title('Types of Shows', fontsize=20)
```
